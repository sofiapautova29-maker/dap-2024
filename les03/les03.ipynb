```python
# %% [markdown]
# ### 3. Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

z = np.arange(11)
print("Исходный массив:")
print(z)

mask = (z > 3) & (z < 8)
z[mask] = -z[mask]

print("\nМассив после изменения знака у элементов от 3 до 8:")
print(z)
```

```python
# %% [markdown]
# ### 4. Найти среднее значение строки матрицы и вычесть его из каждого элемента строки

z = np.random.rand(5, 10)
print("Исходная матрица 5x10:")
print(z)

row_means = z.mean(axis=1, keepdims=True)  # keepdims сохраняет размерность
print("\nСредние значения по строкам:")
print(row_means)

z_centered = z - row_means

print("\nМатрица после центрирования по строкам:")
print(z_centered)

print("\nПроверка (средние после центрирования):")
print(z_centered.mean(axis=1))
```

```python
# %% [markdown]
# ### 5. Поменяйте 1 и 2 строки матрицы

z = np.arange(25).reshape(5,5)
print("Исходная матрица:")
print(z)

row1_copy = z[1].copy()
row2_copy = z[2].copy()

z[1] = row2_copy
z[2] = row1_copy
print("\nМатрица после замены 1 и 2 строк:")
print(z)

z_original = np.arange(25).reshape(5,5)
z_original[[1, 2]] = z_original[[2, 1]]
print("\nАльтернативный способ (проверка):")
print(z_original)
```

```python
# %% [markdown]
# ### 6. Дан массив 10x2, представляющий собой точки в декартовой системе координат, преобразовать в полярную

z = np.random.random((10, 2))
print("Декартовы координаты (10 точек):")
print(z)

# x = r * cos(θ), y = r * sin(θ)
# r = sqrt(x² + y²), θ = arctan2(y, x)

r = np.sqrt(z[:, 0]**2 + z[:, 1]**2)  # Радиус
theta = np.arctan2(z[:, 1], z[:, 0])   # Угол (в радианах)

theta_degrees = np.degrees(theta)

print("\nПолярные координаты:")
print("Радиус (r):", r)
print("Угол в радианах (θ):", theta)
print("Угол в градусах:", theta_degrees)

x_check = r * np.cos(theta)
y_check = r * np.sin(theta)

print("\nПроверка обратного преобразования (декартовы координаты):")
print("x (исходные):", z[:, 0])
print("x (из полярных):", x_check)
print("\ny (исходные):", z[:, 1])
print("y (из полярных):", y_check)
```

```python
# %% [markdown]
# ### 9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением
z = np.array([1, 2, 3, 4, 5])
print("Исходный вектор:")
print(z)

n = len(z)
zeros_between = 3
new_length = n + (n - 1) * zeros_between

result = np.zeros(new_length, dtype=z.dtype)

positions = np.arange(0, new_length, zeros_between + 1)
result[positions] = z
print("\nВектор с тремя нулями между значениями:")
print(result)
print(f"\nПозиции исходных элементов: {positions}")
```

```python
# %% [markdown]
# ### 10. Найти ближайшее к заданному значению число в заданном массиве

z = np.arange(100)
v = np.random.uniform(0, 100)

print(f"Массив: от {z[0]} до {z[-1]}")
print(f"Заданное значение: {v:.6f}")

# Способ 1: используя argmin с абсолютной разницей
idx = np.argmin(np.abs(z - v))
nearest_value = z[idx]

# Способ 2: используя searchsorted для отсортированного массива
idx2 = np.searchsorted(z, v)
if idx2 == 0:
    nearest_value2 = z[0]
elif idx2 == len(z):
    nearest_value2 = z[-1]
else:
    # Сравниваем два ближайших значения
    left = z[idx2 - 1]
    right = z[idx2]
    nearest_value2 = left if abs(v - left) < abs(v - right) else right

print(f"\nБлижайшее значение (способ 1): {nearest_value}")
print(f"Ближайшее значение (способ 2): {nearest_value2}")
print(f"Разница: {abs(v - nearest_value):.6f}")
```

```python
# %% [markdown]
# ### 2. Что будет выведено в консоль и почему?

# %%
print("2. Результаты операций:")
print("=" * 50)

# Операция 1
arr1 = np.array([1, 4, 2, 5, 3], dtype='str')
result1 = arr1[1] + '1'
print(f"1. arr[1] + '1' = '{result1}' (конкатенация строк '4' и '1')")

# Операция 2
arr2 = np.array([1, -4, 2, 0, 3], dtype='bool')
arr3 = np.array([1, -4, 2, 0, 3])
result2 = arr2[3] or arr3[1]
print(f"2. arr2[3] or arr3[1] = {result2} (arr2[3] = 0 -> False, arr3[1] = -4 -> True)")

# Операция 3
result3 = arr2[3] and arr3[1]
print(f"3. arr2[3] and arr3[1] = {result3} (arr2[3] = 0 -> False)")

# Операция 4
arr4 = np.array([1, -4, 2, 0, True], dtype='int64')
result4 = arr4[-1]
print(f"4. arr4[-1] = {result4} (True преобразуется в 1)")

# Операция 5
try:
    arr5 = np.array([1, -4, 2, 0, 1j], dtype='int64')
    result5 = arr5[1]
except Exception as e:
    result5 = f"Ошибка: {type(e).__name__}"
print(f"5. arr5[1] = {result5} (нельзя преобразовать комплексные числа в int64)")

# Операция 6
arr6 = np.array([1, -4, 2, 0, 1j], dtype='str')
result6 = arr6[-1]
print(f"6. arr6[-1] = '{result6}' (комплексное число преобразовано в строку)")

# Операция 7
arr7 = np.array([1, 4, 2, 5, '3'])
result7 = arr7[-1] * 3
print(f"7. arr7[-1] * 3 = '{result7}' (строковое повторение '3' * 3 = '333')")

# %% [markdown]
# **Объяснения:**
# 1. Конкатенация строк '4' и '1' → '41'
# 2. 0 → False, -4 → True, False or True → True
# 3. 0 → False, False and ... → False
# 4. True → 1 в int64
# 5. Ошибка: нельзя преобразовать комплексные числа в int64
# 6. Комплексное число 1j преобразуется в строку '1j'
# 7. Строка '3' * 3 = '333' (повторение строки)
```

```python
# %% [markdown]
# ### 4. Дан двумерный массив X, что будет выведено в консоль?

# %%
print("4. Операции с индексацией массива:")
print("=" * 50)

X = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])

print("Массив X:")
print(X)
print()

# Создаем словарь операций
operations = {
    "X[0,0]": X[0,0],
    "X[-1,-1]": X[-1,-1],
    "X[0:1,-1]": X[0:1,-1],
    "X[0:2,-1]": X[0:2,-1],
    "X[0:2,0:2]": X[0:2,0:2],
    "X[1,1:3]": X[1,1:3],
    "X[1,:]": X[1,:],
    "X[:,:]": X[:,:],
    "X[:]": X[:],
    "X[::2,::2]": X[::2,::2],
    "X[0:3:2,0:3:2]": X[0:3:2,0:3:2],
    "X[::-1,::-1]": X[::-1,::-1]
}

# Выводим результаты
for op, result in operations.items():
    print(f"{op:20} = {result}")

# %% [markdown]
# **Объяснения:**
# - `X[0,0]` → 3 (первый элемент первой строки)
# - `X[-1,-1]` → 7 (последний элемент последней строки)
# - `X[0:1,-1]` → [4] (срез: строка 0, последний столбец)
# - `X[0:2,-1]` → [4, 8] (строки 0-1, последний столбец)
# - `X[0:2,0:2]` → [[3, 5], [7, 6]] (первые 2 строки и столбцы)
# - `X[1,1:3]` → [6, 8] (вторая строка, столбцы 1-2)
# - `X[1,:]` → [7, 6, 8, 8] (вся вторая строка)
# - `X[:,:]` → весь массив (полная копия)
# - `X[:]` → весь массив (альтернативная запись)
# - `X[::2,::2]` → [[3, 2], [1, 7]] (каждая вторая строка и столбец)
# - `X[0:3:2,0:3:2]` → [[3, 2], [1, 7]] (аналогично предыдущему)
# - `X[::-1,::-1]` → [[7, 7, 6, 1], [8, 8, 6, 7], [4, 2, 5, 3]] (массив в обратном порядке по обеим осям)
```

```python
# %% [markdown]
# ### 5. Дан двумерный массив grid, в чем различие между операциями?

# %%
print("5. Различие между np.split с разными осями:")
print("=" * 50)

grid = np.arange(16).reshape((4, 4))
print("Исходный массив grid:")
print(grid)
print()

# Разделение по axis=0 (по строкам)
split_rows = np.split(grid, [2])
print("np.split(grid, [2]) - разделение по строкам:")
for i, part in enumerate(split_rows):
    print(f"Часть {i}:")
    print(part)
    print()

# Разделение по axis=1 (по столбцам)
split_cols = np.split(grid, [2], axis=1)
print("np.split(grid, [2], axis=1) - разделение по столбцам:")
for i, part in enumerate(split_cols):
    print(f"Часть {i}:")
    print(part)
    print()

# %% [markdown]
# **Ответ:**
# - `np.split(grid, [2])` делит массив по строкам (axis=0 по умолчанию) на 3 части: первые 2 строки и остальные
# - `np.split(grid, [2], axis=1)` делит массив по столбцам на 3 части: первые 2 столбца и остальные
```

```python
# %% [markdown]
# ### 6. В чем различие между операциями?

# %%
print("6. Различие между сложением списков и массивов NumPy:")
print("=" * 50)

list1 = [1, 2, 3]
list2 = [3, 2, 1]

arr1 = np.array([1, 2, 3])
arr2 = np.array([3, 2, 1])

# Сложение списков
list_result = list1 + list2

# Сложение массивов NumPy
arr_result = arr1 + arr2

print("Списки:")
print(f"  [1, 2, 3] + [3, 2, 1] = {list_result}")
print("  (конкатенация списков)")

print("\nМассивы NumPy:")
print(f"  np.array([1, 2, 3]) + np.array([3, 2, 1]) = {arr_result}")
print("  (поэлементное сложение)")

# Демонстрация других операций
print("\nДругие операции с массивами NumPy:")
print(f"  arr1 * arr2 = {arr1 * arr2} (поэлементное умножение)")
print(f"  arr1 ** 2 = {arr1 ** 2} (возведение в квадрат)")
print(f"  np.dot(arr1, arr2) = {np.dot(arr1, arr2)} (скалярное произведение)")

# %% [markdown]
# **Ответ:**
# - `[1, 2, 3] + [3, 2, 1]` → `[1, 2, 3, 3, 2, 1]` (конкатенация списков)
# - `np.array([1, 2, 3]) + np.array([3, 2, 1])` → `[4, 4, 4]` (поэлементное сложение)
#
# NumPy выполняет векторные операции, а Python списки — операции над последовательностями.
```

```python
# %% [markdown]
# ### 7. Как в Colab измерить время выполнения кода?

# %%
print("7. Измерение времени выполнения кода в Colab:")
print("=" * 50)

import time

# Способ 1: Использование модуля time
print("Способ 1: Использование time.time()")

start_time = time.time()

# Код для измерения
result = 0
for i in range(1000000):
    result += i

end_time = time.time()
print(f"Время выполнения: {end_time - start_time:.6f} секунд")
print(f"Результат: {result}")
print()

# Способ 2: Использование магической команды %timeit в Colab
print("Способ 2: Использование магических команд Colab")
print("В Colab можно использовать:")
print("  %time - измеряет время однократного выполнения")
print("  %timeit - автоматически выбирает количество повторений")
print("  %%timeit - для измерения времени выполнения всей ячейки")
print()

# Способ 3: Использование time.perf_counter() для большей точности
print("Способ 3: Использование time.perf_counter()")

start = time.perf_counter()

# Код для измерения
arr = np.random.rand(1000, 1000)
result = np.sum(arr)

end = time.perf_counter()
print(f"Время выполнения: {end - start:.6f} секунд")
print(f"Результат: {result:.2f}")
```

```python
# %% [markdown]
# ### 8. Дан массив X, каковы результаты операций?

# %%
print("8. Результаты операций суммирования:")
print("=" * 50)

M = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])

print("Массив M:")
print(M)
print()

# Создаем словарь операций
operations = {
    "np.sum(M)": np.sum(M),
    "np.sum(M, axis=1)": np.sum(M, axis=1),
    "np.sum(M, axis=0)": np.sum(M, axis=0),
    "np.sum(M, 1)": np.sum(M, 1),  # то же что axis=1
}

print("Результаты операций NumPy:")
for op, result in operations.items():
    print(f"{op:25} = {result}")

print("\nРезультаты операций со встроенной функцией sum():")
print("sum(M) =", sum(M))
print("  (суммирует по первому измерению, axis=0)")

# Попытка sum(M, 1) вызовет ошибку, так как у встроенной sum() нет параметра axis
try:
    result = sum(M, 1)
    print(f"sum(M, 1) = {result}")
except TypeError as e:
    print(f"sum(M, 1) вызывает ошибку: {e}")

# %% [markdown]
# **Ответы:**
# - `np.sum(M)` → 58 (сумма всех элементов)
# - `np.sum(M, axis=1)` → [14, 29, 21] (сумма по строкам)
# - `np.sum(M, axis=0)` → [11, 17, 17, 19] (сумма по столбцам)
# - `np.sum(M, axis=2)` → ошибка, т.к. массив 2D
# - `np.sum(M, 1)` → [14, 29, 21] (то же что axis=1)
# - `sum(M)` → [11, 17, 17, 19] (сумма по первому измерению, axis=0)
# - `sum(M, 1)` → ошибка, у встроенной sum() нет параметра axis
```

```python
# %% [markdown]
# ### 9. Дан массив X, что будет выведено в консоль?

# %%
print("9. Булева индексация в NumPy:")
print("=" * 50)

X = np.array([1, 2, 3, 4])

print("Массив X:", X)
print()

# Булевы операции
print("Булевы операции:")
print(f"X > 3  = {X > 3}  (элементы больше 3)")
print(f"X < 3  = {X < 3}  (элементы меньше 3)")
print(f"X == 3 = {X == 3} (элементы равные 3)")
print(f"X != 3 = {X != 3} (элементы не равные 3)")
print()

# Булева индексация
print("Булева индексация (фильтрация):")
print(f"X[X < 3]  = {X[X < 3]}  (элементы меньше 3)")
print(f"X[X > 3]  = {X[X > 3]}  (элементы больше 3)")
print(f"X[X != 3] = {X[X != 3]} (элементы не равные 3)")
print(f"X[X <= 3] = {X[X <= 3]} (элементы меньше или равные 3)")
print()

# Комбинированные условия
print("Комбинированные условия:")
mask = (X > 1) & (X < 4)  # Элементы между 1 и 4 (не включая границы)
print(f"(X > 1) & (X < 4) = {mask}")
print(f"X[(X > 1) & (X < 4)] = {X[mask]}")
```

```python
# %% [markdown]
# ### 10. В чем преимущество массивов NumPy по сравнению с обычными списками?

# %%
print("10. Преимущества NumPy перед обычными списками:")
print("=" * 50)

import sys

# Создаем тестовые данные
list_data = list(range(1000000))
numpy_data = np.arange(1000000)

# 1. Эффективность памяти
print("1. Эффективность памяти:")
print(f"   Размер списка: {sys.getsizeof(list_data):,} байт")
print(f"   Размер массива NumPy: {numpy_data.nbytes:,} байт")
print(f"   Отношение: {sys.getsizeof(list_data) / numpy_data.nbytes:.1f}x")
print()

# 2. Скорость операций
print("2. Скорость операций:")
import time

# Операции со списком
start = time.time()
list_sum = sum(list_data)
list_time = time.time() - start

# Операции с массивом NumPy
start = time.time()
numpy_sum = np.sum(numpy_data)
numpy_time = time.time() - start

print(f"   Суммирование списка: {list_time:.6f} сек")
print(f"   Суммирование массива: {numpy_time:.6f} сек")
print(f"   Ускорение: {list_time / numpy_time:.1f}x")
print()

# 3. Векторные операции
print("3. Векторные операции:")
print("   Списки требуют циклов:")
list_squared = [x**2 for x in range(5)]
print(f"   [x**2 for x in range(5)] = {list_squared}")

print("\n   Массивы NumPy поддерживают векторные операции:")
arr_squared = np.arange(5)**2
print(f"   np.arange(5)**2 = {arr_squared}")
print()

# 4. Многомерные массивы и операции
print("4. Многомерные массивы и операции:")
matrix = np.random.rand(3, 3)
print("   Создание и операции с матрицами 3x3:")
print(matrix)
print(f"\n   Среднее по столбцам: {matrix.mean(axis=0)}")
print(f"   Среднее по строкам: {matrix.mean(axis=1)}")
print()

# 5. Широковещание (broadcasting)
print("5. Широковещание (broadcasting):")
arr = np.array([1, 2, 3, 4, 5])
print(f"   Массив: {arr}")
print(f"   arr + 10 = {arr + 10}")
print(f"   arr * 2 = {arr * 2}")
print(f"   1 / (arr + 1) = {1 / (arr + 1)}")
print()

# 6. Математические функции
print("6. Встроенные математические функции:")
print(f"   np.sin(arr) = {np.sin(arr)}")
print(f"   np.exp(arr) = {np.exp(arr)}")
print(f"   np.log(arr + 1) = {np.log(arr + 1)}")

# %% [markdown]
# **Ответ:**
#
# **Преимущества NumPy:**
# 1. **Эффективность памяти**: Массивы NumPy занимают меньше памяти, так как хранят данные в непрерывном блоке памяти
# 2. **Скорость**: Операции выполняются быстрее благодаря:
#    - Векторизованным операциям на C/C++
#    - Отсутствию проверок типов во время выполнения
#    - Кэш-дружественной организации данных
# 3. **Удобные операции**: Векторные и матричные операции без явных циклов
# 4. **Широковещание**: Автоматическое преобразование массивов разных размеров
# 5. **Богатый функционал**: Встроенные математические, статистические, линейно-алгебраические функции
# 6. **Интеграция**: Совместимость с другими научными библиотеками (SciPy, pandas, matplotlib)
#
# **Обусловлено это преимущество:**
# - Хранением данных в непрерывных блоках памяти
# - Реализацией на C/C++ с оптимизированными алгоритмами
# - Использованием векторных инструкций процессора (SIMD)
# - Минимальными накладными расходами
```
