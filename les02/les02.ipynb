```python
# Лабораторная работа №2
# Основы Python
# Выполнила: Паутова Софья
# Группа: УСИС-25-
# Решение задач с LeetCode

# Задача №1. Two Sum
# Ссылка: https://leetcode.com/problems/two-sum/
# Условие: Найти индексы двух чисел в массиве, которые в сумме дают заданное значение.

class Solution1(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        num_dict = {}

        for i, num in enumerate(nums):
            complement = target - num

            if complement in num_dict:
                return [num_dict[complement], i]

            num_dict[num] = i

        return []

# Тестирование задачи №1
solution1 = Solution1()
# Пример 1
nums1 = [2, 7, 11, 15]
target1 = 9
result1 = solution1.twoSum(nums1, target1)
print(f"Пример 1: nums={nums1}, target={target1} -> {result1}")

# Пример 2
nums2 = [3, 2, 4]
target2 = 6
result2 = solution1.twoSum(nums2, target2)
print(f"Пример 2: nums={nums2}, target={target2} -> {result2}")

# Пример 3
nums3 = [3, 3]
target3 = 6
result3 = solution1.twoSum(nums3, target3)
print(f"Пример 3: nums={nums3}, target={target3} -> {result3}")
```

```python
# Задача №2. Palindrome Number
# Ссылка: https://leetcode.com/problems/palindrome-number/description/
# Условие: Определить, является ли число палиндромом (читается одинаково слева направо и справа налево).
class Solution2(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        # Отрицательные числа не могут быть палиндромами
        if x < 0:
            return False

        # Числа, оканчивающиеся на 0 (кроме 0) не могут быть палиндромами
        if x % 10 == 0 and x != 0:
            return False

        # Более эффективный подход: переворачиваем только половину числа
        reversed_half = 0
        original = x

        while x > reversed_half:
            reversed_half = reversed_half * 10 + x % 10
            x //= 10

        # Для четного количества цифр сравниваем x == reversed_half
        # Для нечетного - отбрасываем среднюю цифру в reversed_half
        return x == reversed_half or x == reversed_half // 10

# Тестирование решения для задачи №2
solution2 = Solution2()

test_cases = [121, -121, 10, 12321, 0, 1234321, 123]
print("Задача 2: Проверка чисел на палиндром")
for num in test_cases:
    result = solution2.isPalindrome(num)
    print(f"Число {num}: {'палиндром' if result else 'не палиндром'}")
```

```python
# Задача №3. Roman to Integer
# Ссылка: https://leetcode.com/problems/roman-to-integer/description/
# Условие: Преобразовать римское число в целое.
class Solution3(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        # Словарь значений римских цифр
        roman_values = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        total = 0
        prev_value = 0

        # Проходим по строке справа налево
        for char in reversed(s):
            current_value = roman_values[char]

            # Если текущее значение меньше предыдущего, вычитаем его
            if current_value < prev_value:
                total -= current_value
            else:
                total += current_value
            prev_value = current_value
        return total
# Тестирование решения для задачи №3
solution3 = Solution3()

test_cases = ["III", "IV", "IX", "LVIII", "MCMXCIV", "XXVII"]
print("Задача 3: Преобразование римских чисел в целые")
for roman in test_cases:
    result = solution3.romanToInt(roman)
    print(f"{roman} = {result}")
```

```python
#  Задача №4. Longest Common Prefix
# Ссылка: https://leetcode.com/problems/longest-common-prefix/description/
# Условие: Найти самый длинный общий префикс среди массива строк.
class Solution4(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""

        # Берем первую строку как основу для сравнения
        prefix = strs[0]

        for string in strs[1:]:
            # Сравниваем текущий префикс с каждой строкой
            while not string.startswith(prefix):
                # Укорачиваем префикс, пока не найдем совпадение
                prefix = prefix[:-1]
                if not prefix:
                    return ""
        return prefix

# Тестирование решения для задачи №4
solution4 = Solution4()

test_cases = [
    ["flower", "flow", "flight"],
    ["dog", "racecar", "car"],
    ["interspecies", "interstellar", "interstate"],
    ["", "test"],
    ["single"],
    []
]

print("Задача 4: Поиск самого длинного общего префикса")
for i, strs in enumerate(test_cases, 1):
    result = solution4.longestCommonPrefix(strs)
    print(f"Пример {i}: {strs} -> '{result}'")
```

```python
# Задача №5. Merge Two Sorted Lists
# Ссылка: https://leetcode.com/problems/merge-two-sorted-lists/description/
# Условие: Объединить два отсортированных связных списка в один отсортированный список.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    @staticmethod
    def create_from_list(values):
        """Создает связный список из списка значений"""
        if not values:
            return None

        head = ListNode(values[0])
        current = head

        for val in values[1:]:
            current.next = ListNode(val)
            current = current.next
        return head

    @staticmethod
    def to_list(node):
        """Преобразует связный список в обычный список"""
        result = []
        current = node

        while current:
            result.append(current.val)
            current = current.next
        return result

class Solution5(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # Создаем фиктивный узел для начала нового списка
        dummy = ListNode()
        current = dummy

        # Проходим по обоим спискам
        while list1 and list2:
            if list1.val <= list2.val:
                current.next = list1
                list1 = list1.next
            else:
                current.next = list2
                list2 = list2.next
            current = current.next

        # Добавляем оставшиеся элементы
        current.next = list1 if list1 else list2
        return dummy.next

# Тестирование решения для задачи №5
solution5 = Solution5()

test_cases = [
    ([1, 2, 4], [1, 3, 4]),
    ([], []),
    ([], [0]),
    ([1, 2, 3], [4, 5, 6]),
    ([4, 5, 6], [1, 2, 3])
]

print("Задача 5: Объединение двух отсортированных списков")
for i, (list1_vals, list2_vals) in enumerate(test_cases, 1):
    list1 = ListNode.create_from_list(list1_vals)
    list2 = ListNode.create_from_list(list2_vals)

    merged = solution5.mergeTwoLists(list1, list2)
    result = ListNode.to_list(merged)
    print(f"Пример {i}: {list1_vals} + {list2_vals} -> {result}")
```

```python
# Задача №6. Longest Substring Without Repeating Characters
# Ссылка: https://leetcode.com/problems/longest-substring-without-repeating-characters/description/
# Условие: Найти длину самой длинной подстроки без повторяющихся символов.
class Solution6(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        # Используем sliding window (скользящее окно)
        char_index = {}  # Хранит последний индекс каждого символа
        max_length = 0
        left = 0  # Левая граница окна

        for right in range(len(s)):
            # Если символ уже встречался и его индекс внутри текущего окна
            if s[right] in char_index and char_index[s[right]] >= left:
                # Сдвигаем левую границу окна
                left = char_index[s[right]] + 1

            # Обновляем индекс текущего символа
            char_index[s[right]] = right

            # Обновляем максимальную длину
            max_length = max(max_length, right - left + 1)
        return max_length

# Тестирование решения для задачи №6
solution6 = Solution6()

test_cases = [
    "abcabcbb",
    "bbbbb",
    "pwwkew",
    "",
    "abcdef",
    "aab",
    "dvdf"
]

print("Задача 6: Длина самой длинной подстроки без повторений")
for i, s in enumerate(test_cases, 1):
    result = solution6.lengthOfLongestSubstring(s)
    print(f"Пример {i}: '{s}' -> {result}")
```

```python
# 4. Как определить функцию в Python? Может ли функция возвращать несколько значений?

print("4. Функции в Python:")

# Пример определения функции
def greet(name, greeting="Привет"):
    """Функция приветствия"""
    return f"{greeting}, {name}!"

print(f"Функция greet: {greet('Софья')}")
print(f"С другим приветствием: {greet('Иван', 'Добрый день')}")

# Функция, возвращающая несколько значений
def calculate_stats(numbers):
    """Возвращает минимум, максимум и среднее значение"""
    if not numbers:
        return None, None, None

    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers) / len(numbers)

    # Возвращаем три значения (кортеж)
    return minimum, maximum, average

# Использование функции с несколькими возвращаемыми значениями
nums = [10, 20, 30, 40, 50]
min_val, max_val, avg_val = calculate_stats(nums)
print(f"\nЧисла: {nums}")
print(f"Минимум: {min_val}, Максимум: {max_val}, Среднее: {avg_val:.2f}")

# Ответ:
# Определение функции: Используется ключевое слово `def`.
# Возврат нескольких значений: Да, функция может возвращать несколько значений через кортеж.
```

```python
# 5. Что такое лямбда-функции и для чего они предназначены?
print("5. Лямбда-функции:")

# Обычная функция
def square(x):
    return x ** 2

# Эквивалентная лямбда-функция
square_lambda = lambda x: x ** 2

print(f"square(5) = {square(5)}")
print(f"square_lambda(5) = {square_lambda(5)}")

# Лямбда-функции часто используются с функциями высшего порядка
numbers = [1, 2, 3, 4, 5]

# Использование с map()
squared = list(map(lambda x: x ** 2, numbers))
print(f"\nКвадраты чисел {numbers}: {squared}")

# Использование с filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Четные числа из {numbers}: {even_numbers}")

# Ответ: Лямбда-функции — это анонимные функции, определяемые с помощью ключевого слова `lambda`. Они предназначены для создания простых функций "на лету".
```

```python
# 6. Может ли тело функции содержать определение другой функции?
print("6. Вложенные функции:")

def outer_function(x):
    """Внешняя функция"""
    print(f"  Внешняя функция получила: {x}")

    def inner_function(y):
        """Вложенная функция"""
        return y * 2

    # Вложенные функции могут использовать переменные внешней функции
    def closure_function():
        """Замыкание - использует переменную из внешней функции"""
        return x * 3

    # Использование вложенных функций
    result1 = inner_function(x)
    result2 = closure_function()

    return result1, result2

# Вызов внешней функции
results = outer_function(5)
print(f"  Результаты: {results}")

# Ответ: Да, в Python можно определять функции внутри других функций. Такие функции называются вложенными или локальными.
```

```python
# 7. Как определить класс в Python? Для чего нужна переменная self?

print("7. Классы в Python:")

# Пример определения класса
class Student:
    # Конструктор класса
    def __init__(self, name, group, age):
        # self ссылается на текущий экземпляр
        self.name = name
        self.group = group
        self.age = age
        self.grades = []

    # Метод класса
    def add_grade(self, grade):
        """Добавить оценку"""
        self.grades.append(grade)

    def get_average(self):
        """Вычислить средний балл"""
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

# Создание экземпляров класса
student1 = Student("Софья", "УСИС-25-4", 20)
student1.add_grade(5)
student1.add_grade(4)

print(f"Студент: {student1.name}")
print(f"Группа: {student1.group}")
print(f"Средний балл: {student1.get_average():.2f}")

# %% [markdown]
# **Ответ:**
# **Определение класса:** Используется ключевое слово `class`.
# **self** — это ссылка на текущий экземпляр класса, используется для доступа к атрибутам и методам класса.
```

```python
# 8. Как определить приватный метод? Можно ли вызвать приватный метод за пределами класса?
print("8. Приватные методы:")

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Приватный атрибут

    # Приватный метод
    def __log_transaction(self, message):
        """Записать транзакцию в лог (приватный метод)"""
        print(f"  [Лог] {message}. Баланс: {self.__balance}")

    # Публичный метод
    def deposit(self, amount):
        """Пополнить счет"""
        if amount > 0:
            self.__balance += amount
            self.__log_transaction(f"Пополнение: +{amount}")
            return True
        return False

# Создание и использование объекта
account = BankAccount("Софья", 1000)
account.deposit(500)

# Попытка вызвать приватный метод напрямую
try:
    account.__log_transaction("Тест")
except AttributeError as e:
    print(f"  Ошибка при вызове приватного метода: {e}")

# Ответ:
# Приватный метод в Python определяется с помощью двойного подчеркивания: `__private_method`.
# Технически можно вызвать приватный метод за пределами класса (через `_ClassName__methodName`), но это нарушает инкапсуляцию.
```

```python
# 9. Как определить статический метод класса? Для чего может понадобиться статический метод?
print("9. Статические методы:")

class MathOperations:
    # Статический метод
    @staticmethod
    def add(a, b):
        """Сложение двух чисел"""
        return a + b

    @staticmethod
    def factorial(n):
        """Вычисление факториала"""
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Использование статических методов без создания экземпляра
print(f"MathOperations.add(5, 3) = {MathOperations.add(5, 3)}")
print(f"MathOperations.factorial(5) = {MathOperations.factorial(5)}")

# Ответ:
# Статический метод определяется с помощью декоратора `@staticmethod`. Он не имеет доступа к `self` или `cls`.
# Назначение: Для методов, которые не требуют доступа к атрибутам экземпляра или класса.
```

```python
# 10. Чем staticmethod отличается от classmethod?
print("10. staticmethod vs classmethod:")

class Employee:
    # Атрибут класса
    company = "TechCorp"

    def __init__(self, name):
        self.name = name

    # Статический метод
    @staticmethod
    def calculate_bonus(salary, percentage):
        """Расчет бонуса (не зависит от класса)"""
        return salary * percentage / 100

    # Метод класса
    @classmethod
    def get_company_info(cls):
        """Получить информацию о компании"""
        return f"Компания: {cls.company}"

print(f"Статический метод: {Employee.calculate_bonus(100000, 15)}")
print(f"Метод класса: {Employee.get_company_info()}")

# Ответ:
# @staticmethod — не получает ни `self`, ни `cls`. Это просто функция в пространстве имен класса.
# @classmethod — получает `cls` (ссылку на класс) и может обращаться к атрибутам класса.
```

```python
# 11. Как определить собственный класс исключений Python? Как выбросить и перехватить исключение?
print("11. Пользовательские исключения:")

# Определение собственного класса исключений
class NegativeNumberError(Exception):
    """Исключение для отрицательных чисел"""
    def __init__(self, value):
        self.value = value
        super().__init__(f"Отрицательное число: {value}")

# Функция, которая генерирует исключение
def calculate_square_root(number):
    """Вычисление квадратного корня"""
    if number < 0:
        raise NegativeNumberError(number)
    return number ** 0.5

# Перехват исключения
try:
    result = calculate_square_root(-4)
    print(f"Корень: {result}")
except NegativeNumberError as e:
    print(f"Поймано исключение: {e}")

# Ответ:
# Собственный класс исключений создается путем наследования от `Exception`.
# Выбросить исключение: `raise Исключение("сообщение")`
# Перехватить исключение: `try: ... except Исключение as e: ...`
```

```python
# 12. Что такое кортеж, для чего он используется?
print("12. Кортежи (tuples):")

# Создание кортежей
numbers = (1, 2, 3, 4, 5)
coordinates = (10, 20, 30)

print(f"Кортеж чисел: {numbers}")
print(f"Кортеж координат: {coordinates}")

# Неизменяемость кортежей
try:
    numbers[0] = 10
except TypeError as e:
    print(f"\nКортеж неизменяем: {e}")

# Использование в качестве ключей словаря
student_grades = {
    ("Иван", "Иванов"): 4.5,
    ("Мария", "Петрова"): 4.8
}
print(f"\nКортеж как ключ словаря: {student_grades}")

# Распаковка кортежей
x, y, z = coordinates
print(f"\nРаспаковка: x={x}, y={y}, z={z}")

# Ответ:
# Кортеж (tuple) — это неизменяемая последовательность элементов. Используется для:
# - Хранения набора значений, которые не должны изменяться
# - Использования в качестве ключей словаря (т.к. они неизменяемы)
# - Возврата нескольких значений из функции
# - Эффективного хранения данных (менее ресурсоемок, чем список)
```
