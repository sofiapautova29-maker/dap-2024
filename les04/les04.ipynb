```python
import numpy as np
import pandas as pd
```

```python
# 1. Создайте DataFrame из файла titanic.csv, в качестве индекса используйте столбец "PassengerId"
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv', index_col='PassengerId')
```

```python
# 2. Просмотрите первые 6 строк
print("Первые 6 строк датафрейма:")
df.head(6)
```

```python
# 3. Выведите описание данных
print("Информация о данных:")
df.info()

df.describe()

print("\nОписательная статистика категориальных признаков:")
df.describe(include=['object'])
```

```python
# 4. Сколько мужчин / женщин находилось на борту?
gender_counts = df['Sex'].value_counts()
print(f"Мужчин: {gender_counts['male']}")
print(f"Женщин: {gender_counts['female']}")
print(f"Всего пассажиров: {len(df)}")
```

```python
# 4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса?
class_distribution = df['Pclass'].value_counts().sort_index()
print("Общее распределение:")
for pclass, count in class_distribution.items():
    print(f"Класс {pclass}: {count} пассажиров ({count/len(df)*100:.1f}%)")

print("\nРаспределение по классам для мужчин:")
men_df = df[df['Sex'] == 'male']
men_class_dist = men_df['Pclass'].value_counts().sort_index()
for pclass, count in men_class_dist.items():
    print(f"Класс {pclass}: {count} мужчин ({count/len(men_df)*100:.1f}%)")

print("\nРаспределение по классам для женщин:")
women_df = df[df['Sex'] == 'female']
women_class_dist = women_df['Pclass'].value_counts().sort_index()
for pclass, count in women_class_dist.items():
    print(f"Класс {pclass}: {count} женщин ({count/len(women_df)*100:.1f}%)")

# Мужчины 2-го класса
men_2nd_class = men_class_dist.get(2, 0)
print(f"\nМужчин 2-го класса: {men_2nd_class}")
```

```python
# 5. Каковы медиана и среднеквадратичное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

fare_median = df['Fare'].median()
fare_std = df['Fare'].std()

print("Статистики по стоимости билетов (Fare):")

print(f"Медиана: {fare_median:.2f}")
print(f"Среднеквадратичное отклонение: {fare_std:.2f}")
print(f"Среднее: {df['Fare'].mean():.2f}")
print(f"Минимум: {df['Fare'].min():.2f}")
print(f"Максимум: {df['Fare'].max():.2f}")
print(f"Количество пропусков: {df['Fare'].isna().sum()}")
```

```python
# 6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?
young_mask = df['Age'] < 30
old_mask = df['Age'] > 60

# Подсчитываем выживших в каждой группе
young_survived = df.loc[young_mask, 'Survived'].sum()
young_total = young_mask.sum()
young_survival_rate = young_survived / young_total if young_total > 0 else 0

old_survived = df.loc[old_mask, 'Survived'].sum()
old_total = old_mask.sum()
old_survival_rate = old_survived / old_total if old_total > 0 else 0

print("Сравнение выживаемости по возрастным группам:")

print(f"Пассажиров моложе 30 лет: {young_total}")
print(f"Выживших моложе 30 лет: {young_survived}")
print(f"Доля выживших (<30 лет): {young_survival_rate:.2%}")
print()
print(f"Пассажиров старше 60 лет: {old_total}")
print(f"Выживших старше 60 лет: {old_survived}")
print(f"Доля выживших (>60 лет): {old_survival_rate:.2%}")
print()

# Проверяем утверждение
if young_survival_rate > old_survival_rate:
    print(f"Утверждение ВЕРНО: люди моложе 30 лет выживали чаще ({young_survival_rate:.2%} vs {old_survival_rate:.2%})")
else:
    print(f"Утверждение ЛОЖНО: люди старше 60 лет выживали чаще ({old_survival_rate:.2%} vs {young_survival_rate:.2%})")
```

```python
# 7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?
survival_by_gender = df.groupby('Sex')['Survived'].agg(['sum', 'count'])
survival_by_gender['survival_rate'] = survival_by_gender['sum'] / survival_by_gender['count']

print("Выживаемость по полу:")
for gender, row in survival_by_gender.iterrows():
    print(f"{gender.capitalize()}:")
    print(f"  Всего: {int(row['count'])}")
    print(f"  Выжило: {int(row['sum'])}")
    print(f"  Доля выживших: {row['survival_rate']:.2%}")
    print()

# Проверяем утверждение
women_rate = survival_by_gender.loc['female', 'survival_rate']
men_rate = survival_by_gender.loc['male', 'survival_rate']

if women_rate > men_rate:
    print(f"Утверждение ВЕРНО: женщины выживали чаще мужчин ({women_rate:.2%} vs {men_rate:.2%})")
else:
    print(f"Утверждение ЛОЖНО: мужчины выживали чаще женщин ({men_rate:.2%} vs {women_rate:.2%})")
```

```python
# 8. Найдите самое популярное имя среди пассажиров Титаника мужского пола?
def extract_first_name(name):
    """Извлекает первое имя из полного имени"""
    # Убираем титулы и фамилии
    parts = name.split(', ')
    if len(parts) > 1:
        # Берем часть после фамилии
        name_part = parts[1]
        # Убираем титулы в скобках
        if '(' in name_part and ')' in name_part:
            name_part = name_part.split('(')[1].split(')')[0]
        # Убираем титулы Mr., Mrs., Miss., Master., Dr. и т.д.
        title_removed = name_part.split('. ')[-1] if '. ' in name_part else name_part
        # Берем первое слово (первое имя)
        first_name = title_removed.split(' ')[0]
        return first_name
    return name

# Применяем функцию к именам мужчин
men_df = df[df['Sex'] == 'male'].copy()
men_df['FirstName'] = men_df['Name'].apply(extract_first_name)

# Находим самое популярное имя
name_counts = men_df['FirstName'].value_counts()
print("Топ-10 самых популярных имен среди мужчин:")

for i, (name, count) in enumerate(name_counts.head(10).items(), 1):
    print(f"{i}. {name}: {count} пассажиров")

most_common_name = name_counts.index[0]
most_common_count = name_counts.iloc[0]
print(f"\nСамое популярное имя среди мужчин: '{most_common_name}'")
print(f"Количество пассажиров с этим именем: {most_common_count}")
```

```python
# 9. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания?
# Расчет среднего возраста по полу и классу
age_by_gender_class = df.groupby(['Sex', 'Pclass'])['Age'].mean().unstack()

print("Средний возраст по полу и классу обслуживания:")
print("=" * 50)
print(age_by_gender_class.round(1))

# Проверка утверждений
print("\nПроверка утверждений:")

# 1. В среднем мужчины 1-го класса старше 40 лет
men_1st_class_age = age_by_gender_class.loc['male', 1]
statement1 = men_1st_class_age > 40
print(f"1. Мужчины 1-го класса старше 40 лет: {men_1st_class_age:.1f} лет -> {statement1}")

# 2. В среднем женщины 1-го класса старше 40 лет
women_1st_class_age = age_by_gender_class.loc['female', 1]
statement2 = women_1st_class_age > 40
print(f"2. Женщины 1-го класса старше 40 лет: {women_1st_class_age:.1f} лет -> {statement2}")

# 3. Мужчины всех классов в среднем старше женщин того же класса
men_vs_women = (age_by_gender_class.loc['male'] > age_by_gender_class.loc['female']).all()
print(f"3. Мужчины всех классов старше женщин того же класса: {men_vs_women}")

# 4. В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса
overall_age_by_class = df.groupby('Pclass')['Age'].mean()
statement4 = (overall_age_by_class[1] > overall_age_by_class[2] > overall_age_by_class[3])
print(f"4. Возраст уменьшается с классом (1 > 2 > 3): {statement4}")
print(f"   Средний возраст по классам: 1-й: {overall_age_by_class[1]:.1f}, "
      f"2-й: {overall_age_by_class[2]:.1f}, 3-й: {overall_age_by_class[3]:.1f}")
```

```python
# 10. Сравните возраст у спасенных и у погибших пассажиров. Средний возраст погибших выше, верно?
# Расчет среднего возраста для выживших и погибших
age_by_survival = df.groupby('Survived')['Age'].agg(['mean', 'median', 'count'])
print("Статистики возраста по выживаемости:")
for survived, stats in age_by_survival.iterrows():
    status = 'Выжили' if survived == 1 else 'Погибли'
    print(f"{status}:")
    print(f"  Средний возраст: {stats['mean']:.1f} лет")
    print(f"  Медианный возраст: {stats['median']:.1f} лет")
    print(f"  Количество: {int(stats['count'])} пассажиров")
    print()

# Проверяем утверждение
mean_age_survived = age_by_survival.loc[1, 'mean']
mean_age_died = age_by_survival.loc[0, 'mean']
print(f"Разница в среднем возрасте: {abs(mean_age_died - mean_age_survived):.1f} лет")

if mean_age_died > mean_age_survived:
    print(f"Утверждение ВЕРНО: средний возраст погибших ({mean_age_died:.1f} лет) "
          f"выше, чем у выживших ({mean_age_survived:.1f} лет)")
else:
    print(f"Утверждение ЛОЖНО: средний возраст выживших ({mean_age_survived:.1f} лет) "
          f"выше, чем у погибших ({mean_age_died:.1f} лет)")
```
